```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
import json
from OpenMonkey import *
```

```python
# dataset images directory
root = 'D:/Research/WebMonkey/'
# json annotation file
js = 'E:/Challenges/data/train.json'
```

```python
# load monkey dataset
op = OpenMonkey(js, root)
```

```python
# Each instance in this dataset has a unique id. For example, to visualize the annotations of id 34 and 84
ids = [34, 84]

# Load the paths of images which have id 34 and 84.
imgs = op.loadImgs(ids)

# Load the corresponding bounding boxes (x1, y1, w, h).
bboxs =  op.loadBbox(ids)

# Load the corresponding annotations.
anns = op.loadAnns(ids)
```

```python
# Display the image and monkey species
op.showImgs(imgs)
op.loadSpecsNames(ids)
```

```python
# Draw the bounding boxes on the images
op.showBbox(imgs, bboxs)
```

```python
# Draw the skeletons on the monkey images
op.showAnns(imgs, anns, bboxs=None, keypoints=False)
```

```python
# Draw the skeletons on the monkey images with keypoints and bounding boxes
op.showAnns(imgs, anns, bboxs, keypoints=True)
```

```python
# Crop the bounding boxes
cropped = op.croppedImgs(imgs, bboxs, display=True)
```

```python
# Crop the bounding boxes and generate the keypoints coordinates after cropping
cropped, xy = op.croppedAnns(imgs, anns, bboxs, True)
print("New x coordinates: " + ' '.join([str(i) for i in xy[0]]))
print("New y coordinates: " + ' '.join([str(i) for i in xy[1]]))
```
