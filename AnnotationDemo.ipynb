```python
import cv2
import numpy as np
from IPython.display import Image
from matplotlib import pyplot as plt
import os
import json
```

```python
def DrawBBox(x1, y1, w, h, img, thickness=2):
    x2 = x1 + w
    y2 = y1 + h
    cv2.rectangle(img, (x1,y1), (x2,y2),(0,255,0), thickness)
    return img
```

```python
annDir= 'P:/MinneZoo/Challenges/data'
imgDir= 'P:/MinneZoo/Challenges/images'
dataType='Train' # Train/Val/Test
annFile= os.path.join(annDir, dataType+'.json')
annFile
```

```python
with open(annFile) as json_file:
    data = json.load(json_file)
```

```python
# Take the first instance as an example
ann = data['annotation'][0]

# Display the image
img = cv2.imread(os.path.join(imgDir, dataType, ann['file_name']))
plt.imshow(img)
plt.show()
```

```python
# The bounding box of the monkey [x1, y1, width, height]
ann['bbox']
```

```python
# Draw the bounding box
bbox = ann['bbox']
im = DrawBBox(bbox[0], bbox[1], bbox[2], bbox[3], cv2.imread(img), thickness=2)
plt.imshow(im)
plt.show()
```
